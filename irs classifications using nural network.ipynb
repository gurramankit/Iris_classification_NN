```python
w1 = np.random.randn(1)
b=np.random.randn(1)
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
for i in range(10):
  for index,row in df.iterrows():
    x_i = np.array(row["height"])
    y_i = np.array(row["target"])

    z = np.dot(w1.transpose(),x_i)+b
    y_hat = sigmoid(z)

    j = j + (y_i*np.log(y_hat)-(1-y_i)*np.log(1-y_hat))

    z_grad=y_hat-y_i
    w1_grad= w1_grad + x_i*z_grad
    b_grad= b_grad + z_grad

    j=j/df.shape[0]
    w1_grad= w1_grad/df.shape[0]
    b_grad= b_grad/df.shape[0]

    w1 = w1 - 0.05*w1_grad
    b = b - 0.05*b_grad

    print(f"The total loss after epoch is {j}")
```

```python
import numpy as np

# Initialize w1 as a NumPy array
w1 = np.random.randn(1)
b = np.random.randn()

def sigmoid(x):
  return 1/(1+np.exp(-x))

j = 0
w1_grad = 0
b_grad = 0

for i in range(10):
  for index, row in df.iterrows():
    x_i = np.array(row["height"])
    y_i = np.array(row["target"])

    # Calculate z using the dot product of w1.transpose() and x_i
    z = np.dot(w1.transpose(), x_i) + b
    y_hat = sigmoid(z)

    j = j + (y_i * np.log(y_hat) - (1 - y_i) * np.log(1 - y_hat))

    z_grad = y_hat - y_i
    w1_grad = w1_grad + x_i * z_grad
    b_grad = b_grad + z_grad

    j = j / df.shape[0]
    w1_grad = w1_grad / df.shape[0]
    b_grad = b_grad / df.shape[0]

    w1 = w1 - 0.1 * w1_grad
    b = b - 0.1 * b_grad

    print(f"The total loss after epoch is {j}")
```
